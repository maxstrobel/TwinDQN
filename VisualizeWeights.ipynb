```python
from single_agent import SingleAgent
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
a = SingleAgent('Breakout', pretrained_model='modelParams/Breakout_final.model')
#a = SingleAgent('SpaceInvaders', pretrained_model='modelParams/SpaceInvaders_final.model')
```

```python
conv1 = next(a.net.conv1.parameters()).data.cpu().numpy()
```

```python
for k in range(len(conv1)):
    f_sum = np.zeros_like(conv1[0,0])
    for i, filt in enumerate(conv1[k]):
        plt.subplot(1,10,i+1)
        plt.imshow(filt)
        
        f_sum += filt
    plt.subplot(1,10,6)
    plt.imshow(conv1[k,1]-conv1[k,0]) 
    plt.subplot(1,10,7)
    plt.imshow(conv1[k,2]-conv1[k,1]) 
    plt.subplot(1,10,8)
    plt.imshow(conv1[k,3]-conv1[k,2])     
    
    plt.subplot(1,10,10)
    plt.imshow(conv1[k,3]-conv1[k,0])     
    plt.show()
```
